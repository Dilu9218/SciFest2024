```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
import ipywidgets as widgets
from IPython.display import display, clear_output

# Function to read a WAV file using librosa
def read_wav(file_path):
    try:
        samples, sample_rate = librosa.load(file_path, sr=None)  # Load audio, `sr=None` keeps the original sample rate
        return sample_rate, samples
    except Exception as e:
        print(f"Error loading audio file: {e}")
        return None, None

# Function to plot the FFT of the audio signal
def plot_fft(sample_rate, samples):
    if len(samples) == 0:
        print("Error: No audio data to process.")
        return
    
    N = len(samples)  # Number of samples
    yf = fft(samples)  # Perform FFT
    xf = fftfreq(N, 1 / sample_rate)  # Create frequency bins

    # Plot the absolute values of the FFT result (magnitude spectrum)
    plt.figure(figsize=(12, 6))
    plt.plot(xf[:N // 2], np.abs(yf[:N // 2]))  # Only plot positive frequencies
    plt.xlim(0,2000)
    plt.title("FFT of the Audio Signal")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.show()


wav_files = [
'InstrumentSoundWavFiles/GrandPiano.wav',#steinway
'InstrumentSoundWavFiles/AcousticGuitar.wav',
'InstrumentSoundWavFiles/Tuba.wav',
'InstrumentSoundWavFiles/FluteSolo.wav',
'InstrumentSoundWavFiles/Harp.wav',
'InstrumentSoundWavFiles/CathedralOrgan.wav',
'InstrumentSoundWavFiles/FrenchHorn.wav',
'InstrumentSoundWavFiles/Saxophone.wav',
'InstrumentSoundWavFiles/Clarinet.wav',
'InstrumentSoundWavFiles/DrumKitHeavyD5.wav'   #Drumkit FFT is very different, played D5
]

# Create a dropdown widget with the .wav file names
dropdown = widgets.Dropdown(
    options=wav_files,
    # value=wav_files,
    description='Select Instrument:',
    disabled=False,
)

# Function to handle dropdown change event
def on_dropdown_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        selected_file = change['new']
        clear_output()
        display(dropdown)
        # sample_rate, samples = read_wav(selected_file)
        # Add any processing or plotting here if needed after loading the file
        # plot_fft(sample_rate, samples)
        sample_rate, samples = read_wav(selected_file)
        plot_fft(sample_rate, samples)


# Attach the event handler to the dropdown
dropdown.observe(on_dropdown_change)

# Display the dropdown
display(dropdown)
```
